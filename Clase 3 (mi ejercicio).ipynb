```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

df=pd.read_csv('dataset_olympics.csv')
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Age', 'Weight', 'Height'], inplace= True)
df['Medal'].fillna(value="NA", inplace=True)
```

```python
df.isnull().sum()
```

```python
df.groupby(['Sport']).size() #separa o fragmenta variables ya sea discreta o continua
```

```python
df.groupby(['Season']).size() #separa o fragmenta variables ya sea discreta o continua
```

```python
df['Medal'].value_counts()
```

```python
df['Sex'].value_counts(normalize=True)
```

```python
(df['Sex'].value_counts(normalize=True))*100
```

```python
pd.crosstab(index=df['Sex'],columns=df['Medal'],margins=True)
```

```python
pd.crosstab(index=df['Season'],columns=df['Sex']).apply(lambda r:r/r.sum())
```

```python
pd.crosstab(index=df['Sex'],columns=df['Medal'],margins=True).apply(lambda r:r/r.sum()*100)
```

```python
df['Medal'].value_counts().plot(kind='bar', title='Ganadores de medallas')
```

```python
df['Medal'].value_counts(normalize=True).plot(kind='bar', title='Ganadores de medallas')
```

```python
df['Medal'].value_counts().plot(kind='pie', autopct='%.2f', title='Ganadores de medallas')
```

```python
pd.crosstab(index=df['Medal'],columns=df['Sex']).plot(kind='bar')
```

```python
pd.crosstab(index=df['Medal'],columns=df['Sex']).plot(kind='bar', stacked=True) #Stacked es que se ponen juntos
```

```python
df['Medal'].value_counts().plot(kind='pie', autopct='%.2f', title='Ganadores de medallas')
```
