```python
df=pd.read_csv('adult_mejorado.csv')
df.isnull().sum()
```

```python
#No se tiene que hacer una limpieza ya que no tenemos ningún valor nulo
#Ahora bien, lo que haré a continuación es determinar 
#el número de personas por cada estado civil, para esto utilizaré la función groupby
```

```python
df['marital-status'].value_counts()
```

```python
#Esto es frecuencia absoluta, lo pasaré a frecuencia relativa con la función normalize
```

```python
(df['marital-status'].value_counts(normalize=True))*100 #Porcentaje de persona por cada estado civil
```

```python
((df['marital-status'].value_counts(normalize=True))*100).plot(kind='bar', title='Estado civil de cada persona')
```

```python
df['marital-status'].value_counts().plot(kind='pie', autopct='%.2f', title='Estado civil de cada persona')
```

```python
#Lo que puedo argumentar es que el estadp civil que mas destaca es el de matrimonio por el civil y el no estár casados. 
#Pero igual puedo argumentar que se me hace mas atractivo de ver la gráfica de barras ya que está ordenado y no tan aglomerado
#Pero lo que si me gusta de la gráfica de pay es que esta tiene los porcentajes colocados, lo que es mas ergonómico a la vista
#Y al momento de analizar la gráfica.
```

```python
pd.crosstab(index=df['workclass'],columns=df['sex'],margins=True)
#Antes de grpaficas, lo que quiero es sacar los totales de tipo de trabajo y género y posteriormente comparar con los 
#resultados graficados.
```

```python
pd.crosstab(index=df['workclass'],columns=df['sex']).apply(lambda r:r/r.sum()*100)
```

```python
pd.crosstab(index=df['workclass'],columns=df['sex']).plot(kind='bar', title='Género en cada trabajo')
#En esta gráfica quiero representar que, dentro de todos los trabajos, que género hay mas, y qué trabajo es donde 
#Mas personas están.
```
