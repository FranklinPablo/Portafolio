# Transportation SA

```python
# Define the number of suppliers and customers
I = 10
J = 15

# Generate random values for ai and bj within a range
ai_values = np.random.randint(100, 1000, size=I)
bj_values = np.random.randint(50, 500, size=J)

# Define the objective function to minimize cost
def objective_function(solution):
    # Define your cost function here using the solution matrix
    # Calculate the total cost
    total_cost = np.sum(solution)  # Example: Total cost is the sum of all elements in the solution matrix
    return total_cost

# Function to generate a random initial solution matrix
def generate_initial_solution():
    # Generate an initial random solution matrix
    # Initialize a matrix with zeros of size (I, J)
    initial_solution = np.zeros((I, J))
    # Randomly assign values in the matrix respecting ai and bj constraints
    # Implement your logic to assign values based on constraints (ai_values and bj_values)
    # Example: Randomly fill the matrix respecting constraints
    for i in range(I):
        for j in range(J):
            initial_solution[i][j] = random.randint(0, min(ai_values[i], bj_values[j]))
    return initial_solution

# Simulated Annealing parameters
T = 1000  # Initial temperature
alpha = 0.9  # Cooling rate
max_iterations = 100  # Maximum iterations

# Simulated Annealing algorithm
def simulated_annealing():
    global T
    current_solution = generate_initial_solution()
    best_solution = current_solution.copy()
    best_cost = objective_function(best_solution)
    iteration_combinations = []
    def perturb_solution(solution):
        # Implement logic to perturb the current solution
        # Example: Randomly modify a value in the solution matrix
        new_solution = solution.copy()
        i = random.randint(0, I - 1)  # Select a random supplier index
        j = random.randint(0, J - 1)  # Select a random customer index
        new_solution[i][j] = random.randint(0, min(ai_values[i], bj_values[j]))  # Modify the selected value
        return new_solution

    for iteration in range(max_iterations):
        T *= alpha  # Cooling schedule

        # Generate a new solution by making a small change to the current solution
        # Implement your logic to generate a neighboring solution
        new_solution = perturb_solution(current_solution)

        # Calculate the cost of the new solution
        new_cost = objective_function(new_solution)

        # Calculate the change in cost
        delta_cost = new_cost - best_cost

        # Check if the new solution is better or if it's accepted probabilistically
        if delta_cost < 0 or random.random() < np.exp(-delta_cost / T):
            current_solution = new_solution.copy()
            if new_cost < best_cost:
                best_solution = new_solution.copy()
                best_cost = new_cost

        # Store iteration and best cost for graphing
        iteration_combinations.append((iteration, best_cost))

    return iteration_combinations, best_solution

# Execute simulated annealing algorithm
iteration_combinations, best_solution = simulated_annealing()

# Print the iteration combinations and the best solution found
print("Iteration combinations:")
for iteration, cost in iteration_combinations:
    print(f"Iteration: {iteration}, Best Cost: {cost}")

print("\nBest Solution:")
print(best_solution)
print("Best Cost:", objective_function(best_solution))

# Plot the graph of iteration vs. best cost
iterations, costs = zip(*iteration_combinations)
plt.figure(figsize=(8, 6))
plt.plot(iterations, costs)
plt.xlabel('Iterations')
plt.ylabel('Best Cost')
plt.title('Simulated Annealing: Iteration vs. Best Cost')
#plt.yscale('log')  # Setting logarithmic scale for the y-axis
plt.grid(True)
plt.show()
```

```python
# Define the number of suppliers and customers
I = 10
J = 15

# Generate random values for ai and bj within a range
ai_values = np.random.randint(100, 1000, size=I)
bj_values = np.random.randint(50, 500, size=J)

# Define the objective function to minimize cost
def objective_function(solution):
    # Calculate the total cost of the solution matrix
    total_cost = np.sum(solution)  # Example: Total cost is the sum of all elements in the solution matrix
    return total_cost

# Generate initial population
def generate_population(population_size):
    population = []
    for _ in range(population_size):
        # Generate a random solution respecting constraints and add to the population
        solution = np.random.randint(0, np.minimum(ai_values[:, np.newaxis], bj_values), size=(I, J))
        population.append(solution)
    return population

# Genetic Algorithm parameters
max_iterations = 100
Pr = 60  # Probability of selection for reproduction
Pc = 40  # Probability of crossover
Pm = 10  # Probability of mutation
population_size = 100

# Implement GA operators: selection, crossover, mutation, replacement

# Run Genetic Algorithm
def genetic_algorithm():
    population = generate_population(population_size)
    best_solution = None
    best_cost = float('inf')
    iteration_combinations = []

    for iteration in range(max_iterations):
        
        # Example code: Random selection of individuals for reproduction (can be replaced with more sophisticated methods)
        selected_parents = random.choices(population, k=int(Pr * population_size))

        # Example code: Single-point crossover (can be replaced with other crossover methods)
        for i in range(0, len(selected_parents), 2):
            crossover_point = random.randint(1, I - 1)
            parent1, parent2 = selected_parents[i], selected_parents[i + 1]
            child1 = np.vstack((parent1[:crossover_point], parent2[crossover_point:]))
            child2 = np.vstack((parent2[:crossover_point], parent1[crossover_point:]))
            population.extend([child1, child2])

        # Example code: Random mutation (can be replaced with more sophisticated methods)
        for i in range(int(Pm * population_size)):
            idx = random.randint(0, population_size - 1)
            mutation_point = (random.randint(0, I - 1), random.randint(0, J - 1))
            population[idx][mutation_point] = random.randint(0, min(ai_values[mutation_point[0]], bj_values[mutation_point[1]]))

        # Example code: Elitism - select best solutions for the next iteration
        costs = [objective_function(sol) for sol in population]
        sorted_indices = np.argsort(costs)
        population = [population[i] for i in sorted_indices[:population_size]]

        # Store iteration and best cost for graphing
        costs = [objective_function(sol) for sol in population]
        idx = np.argmin(costs)
        current_best_solution = population[idx]
        current_best_cost = costs[idx]
        if current_best_cost < best_cost:
            best_solution = current_best_solution
            best_cost = current_best_cost
        iteration_combinations.append((iteration, best_cost))

    return iteration_combinations, best_solution, best_cost

# Execute genetic algorithm
iteration_combinations, best_solution, best_cost = genetic_algorithm()

# Print the best combination and the best objective function
print("Best Combination:")
print(best_solution)
print("Best Objective Function (Cost):", best_cost)

# Plot the graph of iteration vs. best cost
iterations, costs = zip(*iteration_combinations)
plt.plot(iterations, costs)
plt.xlabel('Iterations')
plt.ylabel('Best Cost')
plt.title('Genetic Algorithm: Iteration vs. Best Cost')
plt.show()
```

# Simulated Annealing for Scheduling

```python
#10 Jobs
#Function to calculate the objective value
def objective_function(sequence, processing_times, due_dates, earliness_penalty, tardiness_penalty):
    total_time = 0
    total_earliness_penalty = 0
    total_tardiness_penalty = 0

    for i, job in enumerate(sequence):
        completion_time = max(total_time, due_dates[job])
        earliness = max(0, due_dates[job] - total_time)
        tardiness = max(0, total_time - due_dates[job])

        total_earliness_penalty += earliness * earliness_penalty
        total_tardiness_penalty += tardiness * tardiness_penalty

        total_time = completion_time + processing_times[job]

    return total_time + total_earliness_penalty + total_tardiness_penalty

# Simulated Annealing Algorithm
def simulated_annealing(processing_times, due_dates, earliness_penalty, tardiness_penalty, initial_temperature, cooling_rate, iterations):
    n_jobs = len(processing_times)
    current_sequence = np.arange(n_jobs)
    np.random.shuffle(current_sequence)
    current_objective = objective_function(current_sequence, processing_times, due_dates, earliness_penalty, tardiness_penalty)

    best_sequence = np.copy(current_sequence)
    best_objective = current_objective

    temperatures = [initial_temperature * (cooling_rate ** i) for i in range(iterations)]

    cost_evolution = []  # List to store the cost at each iteration

    for iteration in range(iterations):
        temperature = temperatures[iteration]

        # Generate a neighboring solution by swapping two jobs
        neighbor_sequence = np.copy(current_sequence)
        idx1, idx2 = np.random.choice(n_jobs, size=2, replace=False)
        neighbor_sequence[idx1], neighbor_sequence[idx2] = neighbor_sequence[idx2], neighbor_sequence[idx1]

        neighbor_objective = objective_function(neighbor_sequence, processing_times, due_dates, earliness_penalty, tardiness_penalty)

        # Accept the neighbor if it is better or with a certain probability if it is worse
        if neighbor_objective < current_objective or np.random.rand() < np.exp((current_objective - neighbor_objective) / temperature):
            current_sequence = np.copy(neighbor_sequence)
            current_objective = neighbor_objective

        # Update the best solution
        if current_objective < best_objective:
            best_sequence = np.copy(current_sequence)
            best_objective = current_objective

        cost_evolution.append(best_objective)

    # Plot results
    plt.plot(cost_evolution)
    plt.title('Scheduling with SA')
    plt.xlabel('Iteration')
    plt.ylabel('Production Time')
    plt.grid(True)
    plt.show()

    return best_sequence, best_objective

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
n_jobs = 10
processing_times = np.random.randint(1, 10, size=n_jobs)
due_dates = np.random.randint(10, 20, size=n_jobs)
earliness_penalty = 1
tardiness_penalty = 2
initial_temperature = 1000
cooling_rate = 0.9
iterations = 100

# Run Simulated Annealing
best_sequence, best_objective = simulated_annealing(processing_times, due_dates, earliness_penalty, tardiness_penalty, initial_temperature, cooling_rate, iterations)

# Print results
print("Best Sequence:", best_sequence)
print("Best Objective Value:", best_objective)
```

# Scheduling with Genetic Algorithm

```python
#10 jobs
# Function to calculate the fitness of a schedule
def calculate_fitness(schedule, processing_times, due_dates, earliness_penalty, tardiness_penalty):
    completion_times = np.cumsum(schedule[:, :15], axis=1)
    tardiness = np.maximum(0, completion_times - due_dates)
    earliness = np.maximum(0, due_dates - completion_times)
    total_tardiness_penalty = np.sum(tardiness * tardiness_penalty)
    total_earliness_penalty = np.sum(earliness * earliness_penalty)
    total_time = np.max(completion_times[:, -1])
    fitness = total_time + total_tardiness_penalty + total_earliness_penalty
    return fitness

# Genetic Algorithm Function
def genetic_algorithm(num_iterations, num_population, num_jobs, prob_reproduction, prob_crossover, prob_mutation):
    # Initialize parameters
    processing_times = np.random.randint(1, 20, size=(num_population, num_jobs))
    due_dates = np.random.randint(20, 40, size=(num_population, num_jobs))
    earliness_penalty = np.ones(num_jobs) * 2  # Constant earliness penalty for all jobs
    tardiness_penalty = np.random.randint(1, 5, size=num_jobs)

    best_schedule = None
    best_fitness = float('inf')
    all_fitness = []

    for iteration in range(num_iterations):
        # Evaluate fitness for each individual in the population
        fitness_values = np.zeros(num_population)
        for i in range(num_population):
            fitness_values[i] = calculate_fitness(
                processing_times[i:i+1, :],
                processing_times[i:i+1, :],
                due_dates[i:i+1, :],
                earliness_penalty,
                tardiness_penalty
            )

        # Select parents for reproduction using roulette wheel selection
        total_fitness = np.sum(fitness_values)
        probabilities = fitness_values / total_fitness
        selected_indices = np.random.choice(num_population, size=num_population, p=probabilities)

        # Reproduction (crossover and mutation)
        for i in range(0, num_population, 2):
            if np.random.rand() < prob_reproduction:
                # Crossover
                crossover_point = np.random.randint(1, num_jobs)
                child1 = np.concatenate([processing_times[selected_indices[i], :crossover_point],
                                         processing_times[selected_indices[i+1], crossover_point:]])
                child2 = np.concatenate([processing_times[selected_indices[i+1], :crossover_point],
                                         processing_times[selected_indices[i], crossover_point:]])

                # Mutation
                if np.random.rand() < prob_mutation:
                    mutation_point = np.random.randint(num_jobs)
                    child1[mutation_point] = np.random.randint(1, 20)
                    child2[mutation_point] = np.random.randint(1, 20)

                # Replace parents with children
                processing_times[selected_indices[i], :] = child1
                processing_times[selected_indices[i+1], :] = child2

        # Find the best schedule and update the best fitness
        current_best_index = np.argmin(fitness_values)
        current_best_schedule = processing_times[current_best_index, :]
        current_best_fitness = fitness_values[current_best_index]

        if current_best_fitness < best_fitness:
            best_schedule = current_best_schedule
            best_fitness = current_best_fitness

        all_fitness.append(best_fitness)

    return best_schedule, best_fitness, all_fitness

# Main function
def main():
    num_iterations = 100
    num_population = 100
    num_jobs = 10
    prob_reproduction = 0.6
    prob_crossover = 0.4
    prob_mutation = 0.1

    best_schedule, best_fitness, all_fitness = genetic_algorithm(
        num_iterations, num_population, num_jobs, prob_reproduction, prob_crossover, prob_mutation
    )

    print("Best Schedule:", best_schedule)
    print("Best Fitness:", best_fitness)

    # Plot the evolution of the best fitness
    plt.plot(all_fitness,)
    plt.xlabel('Iterations')
    plt.ylabel('Best Fitness')
    plt.title('Genetic Algorithm for Job Scheduling')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
#15 jobs
# Function to calculate the fitness of a schedule
def calculate_fitness(schedule, processing_times, due_dates, earliness_penalty, tardiness_penalty):
    completion_times = np.cumsum(schedule[:, :15], axis=1)
    tardiness = np.maximum(0, completion_times - due_dates)
    earliness = np.maximum(0, due_dates - completion_times)
    total_tardiness_penalty = np.sum(tardiness * tardiness_penalty)
    total_earliness_penalty = np.sum(earliness * earliness_penalty)
    total_time = np.max(completion_times[:, -1])
    fitness = total_time + total_tardiness_penalty + total_earliness_penalty
    return fitness

# Genetic Algorithm Function
def genetic_algorithm(num_iterations, num_population, num_jobs, prob_reproduction, prob_crossover, prob_mutation):
    # Initialize parameters
    processing_times = np.random.randint(1, 20, size=(num_population, num_jobs))
    due_dates = np.random.randint(20, 40, size=(num_population, num_jobs))
    earliness_penalty = np.ones(num_jobs) * 2  # Constant earliness penalty for all jobs
    tardiness_penalty = np.random.randint(1, 5, size=num_jobs)

    best_schedule = None
    best_fitness = float('inf')
    all_fitness = []

    for iteration in range(num_iterations):
        # Evaluate fitness for each individual in the population
        fitness_values = np.zeros(num_population)
        for i in range(num_population):
            fitness_values[i] = calculate_fitness(
                processing_times[i:i+1, :],
                processing_times[i:i+1, :],
                due_dates[i:i+1, :],
                earliness_penalty,
                tardiness_penalty
            )

        # Select parents for reproduction using roulette wheel selection
        total_fitness = np.sum(fitness_values)
        probabilities = fitness_values / total_fitness
        selected_indices = np.random.choice(num_population, size=num_population, p=probabilities)

        # Reproduction (crossover and mutation)
        for i in range(0, num_population, 2):
            if np.random.rand() < prob_reproduction:
                # Crossover
                crossover_point = np.random.randint(1, num_jobs)
                child1 = np.concatenate([processing_times[selected_indices[i], :crossover_point],
                                         processing_times[selected_indices[i+1], crossover_point:]])
                child2 = np.concatenate([processing_times[selected_indices[i+1], :crossover_point],
                                         processing_times[selected_indices[i], crossover_point:]])

                # Mutation
                if np.random.rand() < prob_mutation:
                    mutation_point = np.random.randint(num_jobs)
                    child1[mutation_point] = np.random.randint(1, 20)
                    child2[mutation_point] = np.random.randint(1, 20)

                # Replace parents with children
                processing_times[selected_indices[i], :] = child1
                processing_times[selected_indices[i+1], :] = child2

        # Find the best schedule and update the best fitness
        current_best_index = np.argmin(fitness_values)
        current_best_schedule = processing_times[current_best_index, :]
        current_best_fitness = fitness_values[current_best_index]

        if current_best_fitness < best_fitness:
            best_schedule = current_best_schedule
            best_fitness = current_best_fitness

        all_fitness.append(best_fitness)

    return best_schedule, best_fitness, all_fitness

# Main function
def main():
    num_iterations = 100
    num_population = 100
    num_jobs = 15
    prob_reproduction = 0.6
    prob_crossover = 0.4
    prob_mutation = 0.1

    best_schedule, best_fitness, all_fitness = genetic_algorithm(
        num_iterations, num_population, num_jobs, prob_reproduction, prob_crossover, prob_mutation
    )

    print("Best Schedule:", best_schedule)
    print("Best Fitness:", best_fitness)

    # Plot the evolution of the best fitness
    plt.plot(all_fitness,)
    plt.xlabel('Iterations')
    plt.ylabel('Best Fitness')
    plt.title('Genetic Algorithm for Job Scheduling')
    plt.show()

if __name__ == "__main__":
    main()
```
